**Introduction to Matplotlib**

```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('_mpl-gallery')
x=np.linspace(0,10,100)
y=4+2*np.sin(x)
fig,a=plt.subplots()
a.plot(x,y,linewidth=3)
a.set(xlim=(0,8),xticks=np.arange(1,8),
      ylim=(0,8),yticks=np.arange(1,8))
plt.show()
```

```python
#lottery game
import random
import matplotlib.pyplot as plt
count1 = 0
count2 = 0
count3 = 0
for i in range(6):
  comp = random.randint(1, 9)
  userInput = int(input("Enter the number: "))
  if userInput < comp:
     count1 += 1
  elif userInput > comp:
    count2 += 1
  else:
   count3 += 1
# Calculate success rates
total_games = count1 + count2 + count3
success_rate_count1 = count1 / total_games
success_rate_count2 = count2 / total_games
success_rate_count3 = count3 / total_games
# Create a bar plot
labels = ['Count1', 'Count2', 'Count3']
success_rates = [success_rate_count1, success_rate_count2, success_rate_count3]
colors = ['blue', 'green', 'orange']
plt.bar(labels, success_rates, color=colors)
plt.xlabel('Counts')
plt.ylabel('Success Rate')
plt.title('Success Rate of Game Counts')
plt.ylim(0, 1) # Set y-axis limits to ensure proper visualization
plt.show()
```
